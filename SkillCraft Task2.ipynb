```python
titanic['age'].fillna(titanic['age'].median(),inplace=True)# Filling the null values in age with median of the data.  
```

```python
titanic['embarked'].fillna(titanic['embarked'].mode()[0],inplace=True)# Filling the null values in embark_town with mode of the data. 
```

```python
titanic.drop(columns=['deck'],inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic['pclass']=titanic['pclass'].astype('category')
```

```python
titanic['pclass'].dtype
```

```python
titanic['survived']=titanic['survived'].astype('bool')
```

# Univariate Analysis

### Survived Column:

```python
sns.countplot(x='survived', data=titanic,palette='pastel' )
plt.title('Survival Count')
plt.ylabel('Count')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

### 2. Age Distribution and Frequency

```python
sns.histplot(titanic['age'].dropna(), kde=True, bins=20, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Inference :
Age group of 25 to 30 is majority

```python
# Filter data where passengers did not survive
unsurvived_data = titanic[titanic['survived'] == False]  # Boolean condition
```

```python
unsurvived_data
```

```python
sns.histplot(unsurvived_data['age'].dropna(), kde=True, bins=20, color='blue')
plt.title('Age Distribution for Unsurvived')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

The most unsurvived ones are in the age group of 25 to 30

```python
# Filter data where passengers survived
survived_data = titanic[titanic['survived'] == True]  # Boolean condition
survived_data
```

```python
sns.histplot(survived_data['age'].dropna(), kde=True, bins=20, color='blue')
plt.title('Age Distribution for Survived')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

The most survived is from 25 to 28(approx)

```python
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title('Sex wise analysis')
plt.show()
```

Males are less survived than Females

### Correlation Analysis

```python
corr = titanic.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Map')
plt.show()
```

### Survival by Class


```python
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Survival by Class')
plt.show()
```

Most survived are 1st class travellers
